```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

client = MongoClient("mongodb://cap.aflohr.com:27017/")
db = client["local"]
collection = db["Term_proj"]
```

```python
cursor = collection.find({"model":"RNN", "num_nodes":120, "lr":0.5}).sort({"epochs":1})
df = pd.DataFrame(list(cursor))
```

```python
print(df)
```

```python
x = df['epochs']
y_2 = df['train_acc']
y = df['test_acc']
```

```python
models = [ 'Forest', 'mlp', 'nb', 'svm', 'tree', 'RNN' ]
datasets = ['KDD_TEST_PLUS', 'KDD_TEST_MINUS']
y_plus = []
y_minus = []
for dataset in datasets:
    print(dataset)
    for model in models:
        print(model)
        cursor = collection.find({"model":model, "dataset":dataset, "isMulti":False}).sort({"test_acc":-1}).limit(1)
        df = pd.DataFrame(list(cursor))
        print(df['test_acc'])
        if dataset == 'KDD_TEST_PLUS':
            y_plus.append(df['test_acc'].iloc[0])
        else:
            y_minus.append(df['test_acc'].iloc[0])
```

```python
print(y_plus, y_minus)
x = np.arange(len(models))  # label locations
width = 0.35  # width of the bars

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, y_plus, width, label='Plus')
bars2 = ax.bar(x + width/2, y_minus, width, label='Minus')

# Labels and titles
ax.set_xlabel('Model')
ax.set_ylabel('Test acc')
ax.set_title('Plus vs Minus Binary')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
y_plus = []
y_minus = []
for dataset in datasets:
    print(dataset)
    for model in models:
        print(model)
        cursor = collection.find({"model":model, "dataset":dataset, "isMulti":True}).sort({"test_acc":-1}).limit(1)
        df = pd.DataFrame(list(cursor))
        print(df['test_acc'])
        if dataset == 'KDD_TEST_PLUS':
            y_plus.append(df['test_acc'].iloc[0])
        else:
            y_minus.append(df['test_acc'].iloc[0])
            
x = np.arange(len(models))  # label locations
width = 0.35  # width of the bars

# Plot
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, y_plus, width, label='Plus')
bars2 = ax.bar(x + width/2, y_minus, width, label='Minus')

# Labels and titles
ax.set_xlabel('Model')
ax.set_ylabel('Test acc')
ax.set_title('Plus vs Minus MultiClass')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```
