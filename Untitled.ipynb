```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt

client = MongoClient("mongodb://cap.aflohr.com:27017/")
db = client["local"]
collection = db["Term_proj"]
```

```python
cursor = collection.find({"model":"RNN", "num_nodes":120, "lr":0.5}).sort({"epochs":1})
df = pd.DataFrame(list(cursor))
```

```python
print(df)
```

```python
x = df['epochs']
y_2 = df['train_acc']
y = df['test_acc']

plt.xlabel('Epochs')
plt.ylabel('Test_Acc')
plt.title('epochs vs Acc')
plt.plot(x,y, label="test")
plt.plot(x, y_2, label="train")
plt.legend()
plt.show()
```

```python
lrs = [.01,.1,.5,.8]
nodes = [20, 60, 80, 120, 240]

for lr in lrs:
    for node in nodes:
        cursor = collection.find({"model":"RNN", 'dataset': 'minus', "num_nodes":node, "lr":lr, "isMulti":False}).sort({"epochs":1})
        cursor_2 = collection.find({"model":"RNN", "num_nodes":node, 'dataset': 'minus', "lr":lr, "isMulti":True}).sort({"epochs":1})
        df = pd.DataFrame(list(cursor))
        df_2 = pd.DataFrame(list(cursor_2))
        x = df['epochs']
        y_2 = df['train_acc']
        y = df['test_acc']
        y_train2 = df_2['train_acc']
        y_test2 = df_2['test_acc']
        plt.xlabel('Epochs')
        plt.ylabel('Test_Acc')
        plt.title('epochs vs Acc ' + str(node) + " " + str(lr))
        plt.plot(x,y, label="test")
        plt.plot(x, y_2, label="train")
        plt.plot(x, y_test2, label="multi_test")
        plt.legend()
        plt.show()
```
